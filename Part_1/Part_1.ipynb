#**ML-style reductions**
In ML, we often compute statistics per feature (column) across samples (rows):

*	mean/std for normalization
*	min/max for scaling
*	sums for loss/gradients
*	covariance for feature relationships

````python
# X: shape (n_samples, n_features)
# Here: 4 samples, 3 features
X = np.random.randn(4, 3)

print("X:\n", X)
print("Shape:", X.shape)  # (4, 3)

 # Global sum: reduces everything to a single number
print("Global sum:", X.sum())

# axis=0: reduce over rows -> gives one value per column (per feature)
print("Sum per feature (axis=0):", X.sum(axis=0))

# axis=1: reduce over columns -> gives one value per row (per sample)
print("Sum per sample (axis=1):", X.sum(axis=1))
````

````python

#Mean, std, min, max (common in preprocessing)
# Feature-wise mean (1 value per feature)
mu = X.mean(axis=0)
print("Mean per feature:", mu)

# Feature-wise std (1 value per feature)
sigma = X.std(axis=0)
print("Std per feature:", sigma)

# Feature-wise min/max
print("Min per feature:", X.min(axis=0))
print("Max per feature:", X.max(axis=0))
````

````python

#standardization z-score

# Standardization: (X - mean) / std
# Broadcasting: mu and sigma (shape (3,)) are applied across all rows
X_std = (X - mu) / (sigma + 1e-8)  # small epsilon to avoid divide-by-zero
print("Standardized X:\n", X_std)

# Check standardized stats (should be ~0 mean and ~1 std per feature)
print("Std mean per feature:", X_std.mean(axis=0))
print("Std std per feature:", X_std.std(axis=0))
````

````python
# Mistake 1: Confusing * with @

X = np.random.randn(5, 3)
w = np.random.randn(3)

# Correct matrix-vector multiplication
y1 = X @ w

# Wrong: elementwise multiply (will broadcast and give (5,3))
y2 = X * w

print("X @ w shape:", y1.shape)  # (5,)
print("X * w shape:", y2.shape)  # (5, 3)
````

````python
'''
চল একদম **intuition + exam-friendly** ভাবে বুঝাই 👌

এই কোডটা আসলে **NumPy তে খুব common একটা mistake** বোঝাচ্ছে।

---

## 🔴 Mistake 1: `*` আর `@` এক জিনিস ভাবা

### 🔹 Data কী আছে আগে দেখি

```python
X = np.random.randn(5, 3)   # 5 rows, 3 columns
w = np.random.randn(3)      # length 3 vector
```

মানে:

* `X` → ৫টা data point, প্রত্যেকটার ৩টা feature
* `w` → weight vector (৩টা weight)

---

## ✅ Correct: `@` (matrix-vector multiplication)

```python
y1 = X @ w
```

এটা গণিতে যা হচ্ছে:

```
(5×3)  dot  (3×1)  =  (5×1)
```

প্রত্যেক row এর সাথে `w` এর **dot product**

Output:

```python
y1.shape  → (5,)
```

👉 একদম **linear model output**
👉 Machine Learning এ যেটা চাই

---

## ❌ Wrong: `*` (element-wise multiplication)

```python
y2 = X * w
```

এখানে কী হচ্ছে?

* `w` এর shape `(3,)`
* NumPy **broadcast** করে:

  ```
  w → (1,3) → (5,3)
  ```
* তারপর element-by-element multiply

Result:

```python
y2.shape → (5, 3)
```

👉 কোনো summation হয়নি
👉 এটা **dot product না**
👉 ML output হিসেবে ভুল

---

## 🔍 ছোট উদাহরণ দিয়ে বুঝি

ধরি:

```
X row = [x1, x2, x3]
w     = [w1, w2, w3]
```

### `@` করলে:

```
x1*w1 + x2*w2 + x3*w3   → 1টা সংখ্যা
```

### `*` করলে:

```
[x1*w1, x2*w2, x3*w3]   → 3টা সংখ্যা
```

---

## 🧠 এক লাইনে বোঝা (exam answer)

> `@` ব্যবহার হয় **matrix / vector multiplication** এর জন্য
> `*` ব্যবহার হয় **element-wise multiplication** এর জন্য
> NumPy তে দুটো এক জিনিস না

---

## 🔑 মনে রাখার trick

* **ML / Linear Algebra** → `@`
* **Same shape element-by-element কাজ** → `*`

চাও তো আমি এটা দিয়ে
👉 **linear regression prediction**
👉 বা **neural network forward pass**
খুব সহজ example এ দেখাতে পারি 😄
''''
````

````python
'''
চল একদম **হাতেকলমে, সংখ্যাসহ** বুঝাই—তাহলে মাথায় গেঁথে যাবে 👍

---

## 🔢 Simple example (no random)

### ধরি:

```python
import numpy as np

X = np.array([
    [1, 2, 3],
    [4, 5, 6]
])          # shape (2,3)

w = np.array([10, 20, 30])   # shape (3,)
```

মানে:

* 2টা data point
* প্রত্যেকটার 3টা feature
* `w` = weight vector

---

## ✅ Correct: `@` (dot product)

```python
y1 = X @ w
print(y1)
print(y1.shape)
```

### কী হচ্ছে ভেতরে?

Row by row হিসাব:

**Row 1**

```
1*10 + 2*20 + 3*30
= 10 + 40 + 90
= 140
```

**Row 2**

```
4*10 + 5*20 + 6*30
= 40 + 100 + 180
= 320
```

### Output:

```
[140 320]
(2,)
```

👉 প্রতিটা row থেকে **একটা output**
👉 এটাই linear regression / NN এ দরকার

---

## ❌ Wrong: `*` (element-wise)

```python
y2 = X * w
print(y2)
print(y2.shape)
```

### কী হচ্ছে?

NumPy `w` কে broadcast করে:

```
w = [10, 20, 30]
↓
[10, 20, 30]
[10, 20, 30]
```

তারপর element-wise multiply:

```
[
 [1*10, 2*20, 3*30],
 [4*10, 5*20, 6*30]
]
```

### Output:

```
[[ 10  40  90]
 [ 40 100 180]]
(2, 3)
```

👉 কোনো যোগ (sum) হয়নি
👉 এটা **dot product না**

---

## 🧠 Visual summary

| Operation       | Code    | Output shape | মানে    |
| --------------- | ------- | ------------ | ------- |
| Matrix × Vector | `X @ w` | `(2,)`       | ✔️ সঠিক |
| Element-wise    | `X * w` | `(2,3)`      | ❌ ভুল   |

---

## 📝 One-line exam answer

> `@` ব্যবহার করলে row-wise dot product হয়,
> `*` ব্যবহার করলে শুধু element-by-element multiplication হয়।

---

চাও তো আমি দেখাতে পারি:

* কেন `X @ w` = prediction
* আর `X * w` কে কিভাবে ঠিক করে sum করলে same result পাওয়া যায়

বললেই হবে 😄
'''
````

````python
'''
চল এটা **খুব পরিষ্কার, ধাপে ধাপে** বুঝাই—এটা NumPy-র সবচেয়ে dangerous broadcasting trap 😄

---

## 🔴 Mistake 2: `(n,)` vs `(n,1)` আলাদা জিনিস

### আগে shape গুলো দেখি

```python
a = np.random.randn(5)      # shape (5,)
b = np.random.randn(5, 1)   # shape (5,1)
```

মানে:

* `a` → **1D vector**

  ```
  [a1 a2 a3 a4 a5]
  ```
* `b` → **column vector**

  ```
  [[b1],
   [b2],
   [b3],
   [b4],
   [b5]]
  ```

---

## ❌ এখন এই লাইনটা দেখো

```python
c = a + b
```

NumPy কী করে?

### 🔹 Broadcasting rule অনুযায়ী:

* `a (5,)` কে ধরে নেয় → `(1,5)`
* `b (5,1)` ঠিকই আছে

এখন shapes:

```
(1,5)
(5,1)
```

এগুলো broadcast হয়ে যায়:

```
(5,5)
```

---

## 🔍 আসলে কী হচ্ছে (visual)

```
a → [a1 a2 a3 a4 a5]
     [a1 a2 a3 a4 a5]
     [a1 a2 a3 a4 a5]
     [a1 a2 a3 a4 a5]
     [a1 a2 a3 a4 a5]

b → [b1]
     [b2]
     [b3]
     [b4]
     [b5]
```

Add করলে:

```
c =
[b1+a1 b1+a2 b1+a3 b1+a4 b1+a5]
[b2+a1 b2+a2 b2+a3 b2+a4 b2+a5]
[b3+a1 b3+a2 b3+a3 b3+a4 b3+a5]
[b4+a1 b4+a2 b4+a3 b4+a4 b4+a5]
[b5+a1 b5+a2 b5+a3 b5+a4 b5+a5]
```

👉 তাই:

```python
c.shape → (5,5)
```

💥 **surprise!**

---

## ✅ ঠিক কী চেয়েছিলে তুমি?

### Case 1️⃣: element-wise add (5 elements)

```python
c = a + b.ravel()   # or b.squeeze()
print(c.shape)     # (5,)
```

### Case 2️⃣: column vector output (5,1)

```python
c = a.reshape(5,1) + b
print(c.shape)     # (5,1)
```

---

## 🧠 Golden rule (exam + real life)

> **NumPy তে `(n,)` ≠ `(n,1)`**
> 1D vector আর column vector এক জিনিস না

---

## 🔑 Memory trick

* `(n,)` → lazy, dangerous 😅
* `(n,1)` → explicit, safe

👉 ML / Linear Algebra এ সবসময় **shape print করে check করো**

চাও তো আমি দেখাতে পারি:

* কেন ML code এ `(n,1)` বেশি safe
* gradient descent এ এই ভুল হলে কী ভয়ংকর bug হয় 😬
'''
````

#*Reshaping & stacking (ML data engineering)*

Why this matters in ML

*	Reshape is used for flattening images, making batches, aligning shapes for matrix math
* Stacking is used when combining datasets, adding features, adding bias terms

````python
# a is a 1D vector with 12 elements
a = np.arange(12)
print("a:", a)
print("a shape:", a.shape)  # (12,)

# Reshape into a matrix: 3 rows, 4 columns
A = a.reshape(3, 4)
print("A:\n", A)
print("A shape:", A.shape)  # (3, 4)

# Transpose swaps rows/columns
print("A.T:\n", A.T)
print("A.T shape:", A.T.shape)  # (4, 3)
````

````python

A = np.ones((3, 2))
B = np.zeros((3, 2))

# Concatenate along axis=1 => append columns (more features)
C_feat = np.concatenate([A, B], axis=1)
print("Feature concat (axis=1):\n", C_feat)
print("Shape:", C_feat.shape)  # (3, 4)

# Concatenate along axis=0 => append rows (more samples)
C_samp = np.concatenate([A, B], axis=0)
print("Sample concat (axis=0):\n", C_samp)
print("Shape:", C_samp.shape)  # (6, 2)

# vstack/hstack are convenience wrappers (mostly for 2D)
print("vstack shape:", np.vstack([A, B]).shape)
print("hstack shape:", np.hstack([A, B]).shape)
````

•	sums for loss/gradients
•	covariance for feature relationships

````python
# X: shape (n_samples, n_features)
# Here: 4 samples, 3 features
X = np.random.randn(4, 3)

print("X:\n", X)
print("Shape:", X.shape)  # (4, 3)

 # Global sum: reduces everything to a single number
print("Global sum:", X.sum())

# axis=0: reduce over rows -> gives one value per column (per feature)
print("Sum per feature (axis=0):", X.sum(axis=0))

# axis=1: reduce over columns -> gives one value per row (per sample)
print("Sum per sample (axis=1):", X.sum(axis=1))
````

````python

#Mean, std, min, max (common in preprocessing)
# Feature-wise mean (1 value per feature)
mu = X.mean(axis=0)
print("Mean per feature:", mu)

# Feature-wise std (1 value per feature)
sigma = X.std(axis=0)
print("Std per feature:", sigma)

# Feature-wise min/max
print("Min per feature:", X.min(axis=0))
print("Max per feature:", X.max(axis=0))
````

````python
#standardization z-score

# Standardization: (X - mean) / std
# Broadcasting: mu and sigma (shape (3,)) are applied across all rows
X_std = (X - mu) / (sigma + 1e-8)  # small epsilon to avoid divide-by-zero
print("Standardized X:\n", X_std)

# Check standardized stats (should be ~0 mean and ~1 std per feature)
print("Std mean per feature:", X_std.mean(axis=0))
print("Std std per feature:", X_std.std(axis=0))
````

Exercise 4 (more ML-relevant)

Create random data X of shape (100, 4). Compute:

1.	per-feature mean
2.	per-feature variance
3.	covariance matrix (feature-feature relationship)

````python
# TODO: Create X with 100 samples and 4 features
X = np.random.randn(100, 4)

# 1) Mean per feature
mu = X.mean(axis=0)

# 2) Variance per feature
var = X.var(axis=0)

# 3) Covariance matrix (4x4), rowvar=False => columns are variables (features)
cov = np.cov(X, rowvar=False)

print("Mean:", mu)
print("Variance:", var)
print("Covariance matrix:\n", cov)
print("Cov shape:", cov.shape)  # should be (4, 4)
````
